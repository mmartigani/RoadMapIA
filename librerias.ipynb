```python
x = np.linspace(-2,2,500)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("square function")
plt.xlabel=("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1
plt.plot(x,y, 'b--', x, y2, 'g')
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1
plt.plot(x,y, 'b--',label="x**2")
plt.plot(x, y2, 'g', label ="x + 1")
plt.legend(loc="best")
plt.show()
```

```python
x = np.linspace(-2,2,500)
y = x**2
y2 = x + 1
plt.subplot(1,2,1)
plt.plot(x,y, 'b--')
plt.subplot(1,2,2)
plt.plot(x, y2, 'g')
plt.show()
```
