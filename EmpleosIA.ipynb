```python
import plotly.express as px
import pandas as pd


jobs_data = {
    'Puesto': [
        'Ingeniero de Inteligencia Artificial',
        'Especialista en Ciberseguridad',
        'Científico de Datos',
        'Ingeniero en Robótica',
        'Arquitecto Cloud',
        'Desarrollador de Automatización',
        'Especialista en Ética de IA',
        'Ingeniero de Software Full Stack',
        'Analista de Datos',
        'Especialista en Experiencia de Usuario (UX)'
    ],
    'Demanda Estimada (2026–2030)': [95, 90, 88, 85, 82, 80, 78, 76, 75, 72]
}

df = pd.DataFrame(jobs_data)

fig = px.bar(
    df,
    x='Demanda Estimada (2026–2030)',
    y='Puesto',
    orientation='h',
    color='Demanda Estimada (2026–2030)',
    color_continuous_scale='turbo',
    title='Ranking de los Empleos Más Demandados (2026–2030)',
    labels={'Demanda Estimada (2026–2030)': 'Demanda (%)'},
    height=600
)


fig.update_layout(
    template='plotly_dark',
    yaxis=dict(autorange='reversed'),
    font=dict(size=14)
)

fig.show()
```
