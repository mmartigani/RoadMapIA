```python
z=pd.read_csv("/content/Datos_Ventas_Tienda.csv")
z.head(3)
```

```python
v=pd.read_csv("/content/Datos_Ventas_Tienda2.csv")
v.head(3)
```

```python
juntos=pd.concat([z,v], ignore_index=True)
juntos.head(7)
```

```python
juntos["Fecha"]=pd.to_datetime(juntos["Fecha"])
juntos.head(3)
```

```python
juntos.info()
```

```python
producto_mas_vendido = juntos.groupby("Producto")["Cantidad"].sum()
producto_mas_vendido = producto_mas_vendido.sort_values(ascending=False)
producto_mas_vendido.head(1)
```

```python
Mes = []
for f in juntos["Fecha"]:
  Mes.append(f.month)
juntos["Mes"] = Mes
juntos.head(3)
```

```python
ventas_por_mes = juntos.groupby("Mes")["Total Venta"].sum().sort_values(ascending=False)
ventas_por_mes.head(1)
```

```python
ventas_por_categoria = juntos.groupby("Producto")["Total Venta"].sum().sort_values(ascending=False)
ventas_por_categoria
```

```python
juntos.to_csv("/content/drive/MyDrive/Resumen_Ventas.csv")
```
